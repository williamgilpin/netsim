```python
def norm_adj(adj):
    """
    given a graph adjacency matrix (with multiple allowed paths
    between a pair of nodes), normalize the rows correctly
    
    Parameters (also attributes)
    ----------
    adj : array
        An adjacency matrix for a graph
    
    
    Returns
    -------
    nadj : array
        a row-normalized stochastic matrix generated from adj

    """
    
    row_sum = (sum(double(adj),axis=1))
    badrow = (row_sum==0.0)
    # fix zero rows via transposition (remove directedness)
    adj[badrow,:] = adj[:,badrow].T
    
    nadj  = (double(adj).T/(row_sum)).T
    nadj[isnan(nadj)] = 0.0
    
    return nadj
    
def renorm1(inmat, k=2, its=1):
    """
    Take a row-normalized transition matrix and 
    k-tuple the columns together 
    NOTE: THE COLUMNS ARE CLUSTERED BY THEIR ORIGINAL SORTING
    
    Parameters (also attributes)
    ----------
    inmat : array
        An graph adjacency matrix to be renormalized
        
    k : int
        The number of states to cluster in each step
    
    its : int
        The number of times to apply the 
        renormalization operation
    
    
    Returns
    -------
    tmat : array
        A renormalized array
    
    """
    
    tmat = copy(inmat)
    for tt in range(its):
        
        sz = tmat.shape

        num_col = sz[1]

        nmat = zeros( [sz[0], ceil(sz[1]/double(k))] )
        nsz = nmat.shape

        for ii in range(nsz[1]-1):
            nmat[:, ii] = sum(tmat[:,(k*ii):(k*(ii+1))], axis=1)

        if (mod(num_col, k) == 0):
            nmat[:, -1] = sum(tmat[:,-k:], axis=1)
        else:
            nmat[:, -1] = sum(tmat[:,-(mod(num_col, k)):], axis=1)
         
        tmat=nmat
        
    return tmat
            

    
def renorm2(inmat, k=2, its=1):
    """
    Uniform weighting assumes local equilibration
    Take a row-normalized transition matrix and 
    k-tuple the rows together 
    NOTE: THE ROWS ARE CLUSTERED BY THEIR ORIGINAL SORTING
    
    Parameters (also attributes)
    ----------
    inmat : array
        An graph adjacency matrix to be renormalized
        
    k : int
        The number of states to cluster in each step
    
    its : int
        The number of times to apply the 
        renormalization operation
    
    
    Returns
    -------
    tmat : array
        A renormalized array
    
    """
    
    tmat = copy(inmat).T
    for tt in range(its):
        
        sz = tmat.shape

        num_col = sz[1]

        nmat = zeros( [sz[0], ceil(sz[1]/double(k))] )
        nsz = nmat.shape

        for ii in range(nsz[1]-1):
            nmat[:, ii] = (1./k)*sum(tmat[:,(k*ii):(k*(ii+1))], axis=1)

        if (mod(num_col, k) == 0):
            nmat[:, -1] = (1./k)*sum(tmat[:,-k:], axis=1)
        else:
            nmat[:, -1] = (1./(mod(num_col, k)))*sum(tmat[:,-(mod(num_col, k)):], axis=1)
         
        tmat = nmat
        
    return tmat.T

def renorm(inmat, k=2, its=1):
    """
    Compute a square renormalization on both the rows and 
    columns of a square array
    """
    return renorm2(renorm1(tmat, k=k, its=its), k=k, its=its)
# qq = rand(300,300)

# imshow(qq)
# figure()
# imshow(renorm1(renorm1(norm_adj(qq), its=4).T,its=4))
# figure()
# imshow(renorm2(renorm1(norm_adj(qq), its=4),its=4))
```

```python
from numpy.random import random_sample

def discrete_dist(vals, weights, nn):
    bins = cumsum(weights)
    return vals[digitize(random_sample(nn), bins)]

def simtraj(trans_mat, tsteps, stt=0):
    """
    Simulation of a trajectory undergoing transitions governed by a jump
    probability matrix
    
    Parameters
    ----------
    
    trans_mat : array
        An NxN matrix of transition probabilities. Current state is 
        given by rows, next state is given by columns
    tsteps : int
        The number of timesteps that elapse
    stt : int
        The state to in which to start the system
        
    Returns
    ----------
    seq : array
        A sequence of states corresponding to a random walk
    
    """
    seq = zeros(tsteps)
    curr = stt
    
    nstates = trans_mat.shape[0]
    states = array([ii for ii in range(nstates)])
    
    for tt in range(tsteps):
        seq[tt] = curr
        weights = copy(trans_mat[curr, :])
        curr = discrete_dist(states, weights, 1)
    return seq

test_traj = simtraj(tmat,3000,stt=0)
plot(test_traj,'-')

figure()
plot(simtraj(renorm2(renorm1(tmat, k=3), k=3),3000,stt=0),'-')
```

```python
def path_ent(tmat, traj):
    """
    Given a trajectory of states sampled from an MSM, calculates the total entropy
    increase of the system along the path
    
    tmat : array
        A transition matrix
    
    traj : array
        A sequence of indices visited
        
    ent : double
        The total stepwise entropy production rate along the path
        
    TODO: RETURN ENTROPY PRODUCED AT EACH STEP, SUM, CUMSUM, etc LATER
    
    """
    traj = array(traj)
    jumps = zip(traj[:-1],traj[1:])
    ent = 0.0
    for jump in jumps:
        ent += log(tmat[jump]/tmat[(jump[1],jump[0])])
    return ent

n_samp = 20
n_rep = 10
len_range = linspace(1,4000,n_samp).astype(int)
mn_ents = zeros(n_samp)
for xx in range(n_rep):
    ents = list()
    for tlen in len_range:
        ttraj = simtraj(tmat,tlen,stt=0)
        ents.append(path_ent(tmat, ttraj)/tlen)
    mn_ents += array(ents)
    
plot(len_range,mn_ents/n_rep)
```
