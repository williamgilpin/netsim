```python
def norm_adj(adj):
    """
    given a graph adjacency matrix (with multiple allowed paths
    between a pair of nodes), normalize the rows correctly
    
    Parameters
    ----------
    adj : array
        An adjacency matrix for a graph
    
    
    Returns
    -------
    nadj : array
        a row-normalized stochastic matrix generated from adj

    """
    
    row_sum = (sum(double(adj),axis=1))
    badrow = (row_sum==0.0)
    # fix zero rows via transposition (remove directedness)
    adj[badrow,:] = adj[:,badrow].T
    
    nadj  = (double(adj).T/(row_sum)).T
    nadj[isnan(nadj)] = 0.0
    
    return nadj
 

# qq = rand(300,300)

# imshow(qq)
```

```python
from numpy.random import random_sample

def discrete_dist(vals, weights, nn=1):
    """
    nn is the number of random values to generate
    """
    bins = cumsum(weights)
    return vals[digitize(random_sample(nn), bins)]

def simtraj(trans_mat, tsteps, stt=0):
    """
    Simulation of a trajectory undergoing transitions governed by a jump
    probability matrix
    
    Parameters
    ----------
    
    trans_mat : array
        An NxN matrix of transition probabilities. Current state is 
        given by rows, next state is given by columns
    tsteps : int
        The number of timesteps that elapse
    stt : int
        The state to in which to start the system
        
    Returns
    ----------
    seq : array
        A sequence of states corresponding to a random walk
    
    """
    seq = zeros(tsteps)
    curr = stt
    
    nstates = trans_mat.shape[0]
    states = array([ii for ii in range(nstates)])
    
    for tt in range(tsteps):
        seq[tt] = curr
        weights = copy(trans_mat[curr, :])
        curr = discrete_dist(states, weights, nn=1)
    return seq

test_traj = simtraj(tmat,3000,stt=0)
plot(test_traj,'-')

figure()
plot(simtraj(renorm(tmat, k=3),3000,stt=0),'-')
```

```python
def path_ent(tmat, traj):
    """
    Given a trajectory of states sampled from an MSM, calculates the total entropy
    increase of the system along the path
    
    tmat : array
        A transition matrix
    
    traj : array
        A sequence of indices visited
        
    ent : double
        The total stepwise entropy production rate along the path
        
    TODO: RETURN ENTROPY PRODUCED AT EACH STEP, SUM, CUMSUM, etc LATER
    
    """
    traj = array(traj)
    jumps = zip(traj[:-1],traj[1:])
    ent = 0.0
    for jump in jumps:
        ent += log(tmat[jump]/tmat[(jump[1],jump[0])])
    return ent

n_samp = 20
n_rep = 10
len_range = linspace(1,4000,n_samp).astype(int)
mn_ents = zeros(n_samp)
for xx in range(n_rep):
    ents = list()
    for tlen in len_range:
        ttraj = simtraj(tmat,tlen,stt=0)
        ents.append(path_ent(tmat, ttraj)/tlen)
    mn_ents += array(ents)
    
plot(len_range,mn_ents/n_rep)
```
